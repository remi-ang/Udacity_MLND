# Data Visualization with Matplotlib

```python
# imports
import numpy as np
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt

matplotlib.style.use('ggplot') # because it looks pretty
```

```python
wheatdata_ff = r'C:\Users\ng35019\Documents\Training\python_for_ds\Module3\Datasets\wheat.data'

df = pd.read_csv(wheatdata_ff)
df.head(10)
```

```python
# create type column
df['type'] = df.wheat_type.astype("category").cat.codes
df.head(10)
```

```python
# draw an histogram

for wtype in df.type.unique():
    df[df.type == wtype].asymmetry.plot.hist(alpha = 0.6)
plt.show()
```

```python
# another dataset
studentdata_ff = r'C:\Users\ng35019\Documents\Training\python_for_ds\Module3\students.data'
df = pd.read_csv(studentdata_ff, index_col=0)
df
```

## Parallel coordinate

```python
from sklearn.datasets import load_iris
from pandas.tools.plotting import parallel_coordinates
```

```python
# load the iris dataset

data = load_iris()
df = pd.DataFrame(data.data, columns=data.feature_names)
df
```

```python
# add target name column
df['target_names'] = [data.target_names[i] for i in data.target]
df
```

```python
# create the parrallel coordinate plot

plt.figure()
parallel_coordinates(df, 'target_names')
plt.show()
```

```python
# Andrew's curves

from pandas.tools.plotting import andrews_curves
```

```python
plt.figure()
andrews_curves(df, 'target_names')
plt.show()
```

```python
# image based plot

df = pd.DataFrame(np.random.randn(1000, 5), columns=['a','b','c','d','e'])
df.corr()
```

```python
plt.imshow(df.corr(), cmap=plt.cm.Blues, interpolation='nearest')
plt.colorbar()
tick_marks = [i for i in range(len(df.columns))]
plt.xticks(tick_marks, df.columns, rotation='vertical')
plt.yticks(tick_marks, df.columns)
plt.show()
```
