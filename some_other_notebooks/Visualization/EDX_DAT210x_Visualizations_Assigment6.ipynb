```python
# imports 

import pandas as pd
from pandas.tools.plotting import andrews_curves
import matplotlib
import matplotlib.pyplot as plt

matplotlib.style.use('ggplot') # because it looks pretty
```

```python
# 1.Loads up the seeds dataset, located at Module3/Datasets/wheat.data into a dataframe

wheatdata_ff = r'C:\Users\ng35019\Documents\Training\python_for_ds\Module3\Datasets\wheat.data'
df = pd.read_csv(wheatdata_ff); df.head(10)
```

```python
# 2.Drop the id column from the dataframe
df1 = df.drop(['id'], axis=1); df1.head(10)
```

```python
# 3.Compute the correlation matrix of your dataframe

corr = df1.corr(); corr
```

```python
help(plt.imshow)
```

```python
# 4.Graph the correlation matrix using imshow or matshow (either or)

plt.figure(figsize=(10,10))
plt.imshow(corr, cmap=plt.cm.Blues, interpolation='nearest', aspect='auto')
plt.colorbar()
tick_marks = [i for i in range(len(df1.columns))]
plt.xticks(tick_marks, df1.columns, rotation='vertical')
plt.yticks(tick_marks, df1.columns)
plt.show()
```
